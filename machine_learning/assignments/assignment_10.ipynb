```python
# QUESTION 3 Create CSV with name & hyperlink from webpage
# Save page source before scraping
import pandas as pd
url = "http://bioguide.congress.gov/biosearch/biosearch1.asp"
response = requests.get(url)

# Save HTML source
with open("congress.html", "w", encoding="utf-8") as f:
    f.write(response.text)

soup = BeautifulSoup(response.text, "html.parser")

names = []
links = []

for a in soup.find_all('a', href=True):
    names.append(a.text.strip())
    links.append(a['href'])

df = pd.DataFrame({
    "Name": names,
    "Link": links
})

df.to_csv("congress_data.csv", index=False)
```

```python
# QUESTION 6 Perform Web Scraping on given HTML
# sample HTML
html_doc = """
<html>
<head><title>Page title</title></head>
<body>
<p id="firstpara" align="center">This is paragraph<b>one</b></p>
<p id="secondpara" align="blah">This is paragraph<b>two</b></p>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, "html.parser")
```

```python
# i) Well-formatted / indented page
print(soup.prettify())
```

```python
# ii) Print <b> tag
print(soup.b)
```

```python
# iii) Print all tags starting with b
for tag in soup.find_all():
    if tag.name.startswith('b'):
        print(tag)
```

```python
# iv) Print text from title and p using list
texts = [soup.title.get_text()] + [p.get_text() for p in soup.find_all('p')]
print(texts)
```

```python
# v) Print text from title and p using dictionary
text_dict = {
    "title": soup.title.get_text(),
    "p": [p.get_text() for p in soup.find_all('p')]
}
print(text_dict)
```

```python
# vi) Print all tag names
tags = set(tag.name for tag in soup.find_all())
print(tags)
```

```python
# vii) Print tags with exactly two attributes
for tag in soup.find_all():
    if len(tag.attrs) == 2:
        print(tag)
```

```python
# viii) Tags with one-character names & no attributes
for tag in soup.find_all():
    if len(tag.name) == 1 and not tag.attrs:
        print(tag)
```

```python
# ix) Tags with align="center"
for tag in soup.find_all(align="center"):
    print(tag)
```

```python
# x) XML â€“ print attributes where name = Alice
xml_doc = '<person name="Bob"><parent rel="mother" name="Alice"></parent></person>'
soup_xml = BeautifulSoup(xml_doc, "xml")

for tag in soup_xml.find_all():
    if tag.get("name") == "Alice":
        print(tag.attrs)
```
