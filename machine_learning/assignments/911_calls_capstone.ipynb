Compute --Create new features &Question 3: What is the most common Reason for a 911 call based on Reason Column? Which comes second

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
df['Reason'].value_counts()
```

Compute --Plot barchart using matplot for911 calls by Reason & Question 4: How can you plot the bars horizontally?

```python
plt.figure(figsize=(6,4))
df['Reason'].value_counts().plot(kind='bar')
plt.title("911 Calls by Reason")
plt.show()
```

```python
# horizontal bars - Use barh() or kind='barh'
plt.figure(figsize=(6,4))
df['Reason'].value_counts().plot(kind='barh')
plt.title("911 Calls by Reason")
plt.show()
```

Do data manipulation & Question 5: Which day got maximum calls for EMS and how many?

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df['Day'] = df['timeStamp'].dt.day_name()

ems = df[df['Reason'] == 'EMS']
ems['Day'].value_counts()

#Friday – 19,938 calls - maximum calls for EMS 
```

Compute --Create a countplot of the Day of Week column with the hue based of the Reason column & Question 6: On which day traffic calls were lowest?

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Day', data=df, hue='Reason')
plt.title("911 Calls by Day and Reason")
plt.legend(loc='upper right')
plt.show()
```

```python
traffic = df[df['Reason'] == 'Traffic']
traffic['Day'].value_counts()
```

```python
# Answer - Sunday (lowest traffic calls)
```

Compute --Create a countplot month wise --Question 7: Which month saw highest calls for fire?

```python
df['Month'] = df['timeStamp'].dt.month

plt.figure(figsize=(10,5))
sns.countplot(x='Month', data=df, hue='Reason')
plt.title("Month wise 911 Calls")
plt.show()
```

Compute --Create Web Map for Traffic Calls & Question 8: Why some areas seem to have lower or almost zero traffic calls? Hint: Zoom the map

A web map visualization can be created using Python libraries such as folium to observe traffic call density. Areas with low or zero calls correspond to regions like airports, parks, and restricted zones where vehicular movement is minimal.

```python
import folium

traffic_df = df[df['Reason'] == 'Traffic']

m = folium.Map(location=[40.13, -75.3], zoom_start=10)

for lat, lng in zip(traffic_df['lat'], traffic_df['lng']):
    if not np.isnan(lat):
        folium.CircleMarker(
            location=[lat, lng],
            radius=1,
            color='blue',
            fill=True
        ).add_to(m)

m
```
