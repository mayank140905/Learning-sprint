15. When the same invitations are sent to many people, the body of the mail does not change. Only the name (and maybe address) needs to be changed.Mail merge is a process  of  doing  this.  Instead  of  writing  each  mail  separately,we can have  a template for the body of the mail and a list of names thatcan be merged.

Create a text file â€œnames.txtâ€ having the names:
Anil
Sunita
Suman
Lokesh
Sumita
John
Johny

Create a text file â€œbody.txtâ€ having the body of the emailas written below:I am going to Delhi. Letâ€™s meet on 7th Jan 2018

```python
import os

# folder name
folder_name = "output_mails"

# create folder if not exists
os.makedirs(folder_name, exist_ok=True)

# read names
with open("output_mails/names.txt", "r") as f:
    names = f.read().splitlines()

# read body
with open("output_mails/body.txt", "r") as f:
    body = f.read()

# create files inside the folder
for name in names:
    file_path = os.path.join(folder_name, f"{name}.txt")
    with open(file_path, "w") as file:
        file.write(f"Hello {name}\n")
        file.write(body)
```
