STEP 0: Import libraries

```python
import pandas as pd
import numpy as np
```

```python
# STEP 1: Read the three CSV files

math = pd.read_csv("data/MathScoreTerm1.csv")
physics = pd.read_csv("data/PhysicsScoreTerm1.csv")
ds = pd.read_csv("data/DSScoreTerm1.csv")
```

```python
# STEP 2: Remove Name and Ethnicity (confidentiality)
for df in [math, physics, ds]:
    df.drop(columns=['Name', 'Ethnicity'], inplace=True, errors='ignore')
```

```python
# STEP 3: Fill missing score data with ZERO
math.fillna(0, inplace=True)
physics.fillna(0, inplace=True)
ds.fillna(0, inplace=True)
```

```python
# STEP 4: Merge the three files
merged_df = math.merge(physics, on='ID') \
                .merge(ds, on='ID')
```

```python
# STEP 5: Convert Sex column (M/F → 1/2)
merged_df['Sex'] = merged_df['Sex'].map({'M': 1, 'F': 2})
```

```python
# STEP 6: Store the data in a new file
merged_df.to_csv("data/ScoreFinal.csv", index=False)
```

```python
# Enhancement 1 (safe version)
if 'Ethnicity' in merged_df.columns:
    merged_df['Ethnicity'] = merged_df['Ethnicity'].astype('category').cat.codes
```

```python
# Enhancement 2: Fill missing score with class average
score_cols = merged_df.select_dtypes(include=np.number).columns

merged_df[score_cols] = merged_df[score_cols].apply(
    lambda x: x.fillna(x.mean())
)
```
