# What is NumPy?

NumPy is a low-level Python library written in C & FORTRAN

Used for high-performance numerical & mathematical computations

Provides a powerful N-dimensional array object

Faster than Python lists because it uses contiguous memory

# Applications of NumPy

Powerful N-dimensional arrays

Sophisticated broadcasting functions

Tools for integrating C/C++ & Fortran code

Useful in Linear Algebra, Fourier Transform, Random Numbers

Backbone of ML / Data Science



# Properties of NumPy Arrays

Homogeneous (same data type)

Stores only numbers (int, float, complex)

Fixed size

Faster computation

```python
Time comparison example
import time

x = range(100000)
y = range(100000)

start = time.time()
c = [a+b for a,b in zip(x,y)]
print(time.time()-start)

a = np.arange(100000)
b = np.arange(100000)

start = time.time()
c = a+b
print(time.time()-start)
```
